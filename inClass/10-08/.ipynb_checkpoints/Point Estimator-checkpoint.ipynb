# Point Estimators

The sample mean is usually not exactly the same as the population mean. Since there might be some bias. 

> Which is why confidence intervals along with point estimates to understand the range in whihc the true parameter falls.


confidence interval (CI) = sample mean +- margin of error

Margin of error is based on your desired confidence level


How to measure the margin of error: z-critical values

Margin of error = z(z-critical value) * standard deviation / sqrt(sample size)

t-critical value is drawn what's known as a t-distribution, if we don't know the sigma. 

a new variable named df = (degree of freedom)

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
data = load_iris()
X = data['data']
y = data['target']
df = pd.DataFrame(X, columns=data['feature_names'])

# Select pairs of variables: sepal length vs sepal width, petal length vs petal width
selected_pairs = df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)', 'petal width (cm)']]

# Calculate the covariance matrix
cov_matrix = selected_pairs.cov()
print("Covariance Matrix:\n", cov_matrix)

# Calculate the correlation matrix
corr_matrix = selected_pairs.corr()
print("Correlation Matrix:\n", corr_matrix)

# Scatter plots with correlation coefficients
def scatter_plot_with_corr(x, y, xlabel, ylabel):
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=x, y=y)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    
    # Compute correlation coefficient
    corr_coef = np.corrcoef(x, y)[0, 1]
    
    # Add correlation coefficient to plot
    plt.title(f"{xlabel} vs {ylabel}\nCorrelation: {corr_coef:.2f}")
    plt.show()

# Plot Sepal length vs Sepal width
scatter_plot_with_corr(df['sepal length (cm)'], df['sepal width (cm)'], 'Sepal Length (cm)', 'Sepal Width (cm)')

# Plot Petal length vs Petal width
scatter_plot_with_corr(df['petal length (cm)'], df['petal width (cm)'], 'Petal Length (cm)', 'Petal Width (cm)')
```
